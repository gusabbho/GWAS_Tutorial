```python
import numpy as np
import matplotlib.pyplot as plt

# Create a sample array
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Plot the data
plt.plot(x, y)
plt.title("Simple Sine Wave Plot")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define labels and descriptions for each column in .ped and .map files
ped_labels = ["Family ID", "Individual ID", "Paternal ID", "Maternal ID", "Sex", "Phenotype", "Genotype Data"]
ped_description = [
    "Unique identifier for each family",
    "Unique identifier for each individual",
    "ID of the father (if known)",
    "ID of the mother (if known)",
    "Sex of the individual (1 = male, 2 = female)",
    "Phenotype (e.g., disease status)",
    "Genotype data for each SNP (e.g., A A, A T, etc.)"
]

map_labels = ["Chromosome", "SNP ID", "Genetic Distance", "Base-pair Position"]
map_description = [
    "Chromosome number where the SNP is located",
    "Unique identifier for each SNP (marker)",
    "Genetic distance in centimorgans (optional)",
    "Physical position on the chromosome (in base pairs)"
]

# Plotting the .ped and .map file structure
fig, ax = plt.subplots(figsize=(10, 8))

# Draw the .ped file structure
for i, (label, desc) in enumerate(zip(ped_labels, ped_description)):
    ax.text(0.1, 0.9 - i * 0.12, f"{label}: {desc}", ha="left", va="center", fontsize=10, color="purple")

# Draw the .map file structure
for i, (label, desc) in enumerate(zip(map_labels, map_description)):
    ax.text(0.6, 0.9 - i * 0.12, f"{label}: {desc}", ha="left", va="center", fontsize=10, color="orange")

# Set titles
ax.text(0.1, 1.0, ".ped File Structure", ha="left", va="center", fontsize=12, color="purple", fontweight="bold")
ax.text(0.6, 1.0, ".map File Structure", ha="left", va="center", fontsize=12, color="orange", fontweight="bold")

# Customize plot appearance
ax.axis("off")
plt.title("PLINK Text File Pair Structure: .ped and .map Files", fontsize=14, fontweight="bold")
plt.show()
```
